# first ML **Model**

# **Load Data**

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

# **Data preparation** **bold text**

## **Data Separation as input and target** *italicized text*

```python
x=df.drop('logS',axis=1)
y=df['logS']
```

## Data Splitting


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=100)
```

# **Model Building**

## **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_msq=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_msq=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
lr_results=pd.DataFrame(['Linear Regression',lr_train_msq,lr_train_r2,lr_test_msq,lr_test_r2]).transpose()
lr_results.columns=['Method','Training MSQ','Training R2','Test MSQ','Test R2']
lr_results
```

## **Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
y_rf_train_msq=mean_squared_error(y_rf_train_pred,y_train)
y_rf_train_r2=r2_score(y_rf_train_pred,y_train)

y_rf_test_msq=mean_squared_error(y_rf_test_pred,y_test)
y_rf_test_r2=r2_score(y_rf_test_pred,y_test)
```

```python
rf_results=pd.DataFrame(['RandomForest',y_rf_train_msq,y_rf_train_r2,y_rf_test_msq,y_rf_test_r2]).transpose()
rf_results.columns=['Method','Training MSQ', 'Training R2','Test MSQ','Test R2']
rf_results
```

#  **Comparision**

```python
df_results=pd.concat([lr_results,rf_results],axis=0).reset_index(drop=True)
df_results
```
